```python
import subprocess
import sys

def run(cmd):
    print("\n>>>", " ".join(cmd))
    subprocess.check_call(cmd)

if __name__ == "__main__":
    py = sys.executable

    run([py, "scripts/train_ie_gan.py"])
    run([py, "scripts/generate_synthetic.py"])

    run([py, "scripts/extract_dinov2_features.py"])
    run([py, "scripts/run_hgwos_feature_selection.py"])
    run([py, "scripts/train_moe.py"])

    print("\nPipeline finished successfully.")
```
